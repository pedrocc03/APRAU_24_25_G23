### Features:

* **Altitude**: A altitude do terreno em metros.
* **Slope_Orientation (Orientação da Inclinação)**: A direção da bússola para a qual o terreno está voltado (em graus).
* **Slope (Inclinação)**: A inclinação em graus.
* **Horizontal_Distance_To_Water (Distância Horizontal até a Água)**: Distância horizontal até a superfície de água mais próxima (em metros).
* **Vertical_Distance_To_Water (Distância Vertical até a Água)**: Distância vertical até a superfície de água mais próxima (em metros).
* **Horizontal_Distance_To_Roadways (Distância Horizontal até Estradas)**: Distância horizontal até a estrada mais próxima, em metros (pode indicar influência de atividades humanas no ambiente).
* **Shadow_Index (Índice de Sombra)**: Índice de sombra medido em três horários do dia (9h, 12h e 15h).
* **Horizontal_Distance_To_Fire_Points (Distância Horizontal até Pontos de Fogo)**: Distância horizontal até os pontos de ignição de incêndios mais próximos (em metros).
* **Canopy_Density (Densidade do Dossel)**: A percentagem da área do terreno coberta por dossel de árvores (medido em percentagem, de 0 a 100).
* **Rainfall_Summer (Precipitação no Verão)**: Precipitação média durante o verão (medida em mm).
* **Rainfall_Winter (Precipitação no Inverno)**: Precipitação média durante o inverno (medida em mm).
* **Wind_Exposure_Level (Nível de Exposição ao Vento)**: Velocidade média do vento durante o inverno (medida em km/h).
* **Soil_Type (Tipo de Solo)**: Designação do tipo de solo. No conjunto de dados existem 40 tipos diferentes de solo.
* **Wilderness_Area (Área Selvagem)**: No conjunto de dados existem 4 tipos diferentes de áreas selvagens.
* **Vegetation_Type (Tipo de Vegetação)**: A variável alvo. Existem 7 classes possíveis, cada uma representando um tipo diferente de vegetação.

```python
# Função para verificar NaNs e Outliers
def check_nans_outliers(df, class_name):
    print(f"Análise de NaNs e Outliers para {class_name}")
    
    # Filtrar para a classe objetivo
    df_class = df[df['Vegetation_Type'] == class_name] 
    
    # Verificar NaNs
    print(f"\n{class_name} - Verificação de NaNs:")
    print(df_class.isnull().sum())
    
    # Selecionar apenas colunas numéricas
    df_numeric = df_class.select_dtypes(include=[np.number])
    
    # Análise de Outliers
    print(f"\n{class_name} - Outliers (nas colunas numéricas):")
    Q1 = df_numeric.quantile(0.25)
    Q3 = df_numeric.quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df_numeric < (Q1 - 1.5 * IQR)) | (df_numeric > (Q3 + 1.5 * IQR)))
    
    # Mostrar o número de outliers em cada coluna
    print(outliers.sum())

# Verificação de NaNs e Outliers para as 3 classes
for class_name in df_all['Vegetation_Type'].unique():
    check_nans_outliers(df_all, class_name)
```

```python
def remover_outliers(df, atributo_objetivo, atributo_preditivo):
    # Criação de uma cópia do dataset 
    df_sem_outliers = df.copy()

    for classe in atributo_objetivo:
        for coluna in atributo_preditivo:
            # Cálculo dos limites para identificar outliers
            Q1 = df_sem_outliers[df_sem_outliers['Vegetation_Type'] == classe][coluna].quantile(0.25)
            Q3 = df_sem_outliers[df_sem_outliers['Vegetation_Type'] == classe][coluna].quantile(0.75)
            IQR = Q3 - Q1

            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            # Identificação e remoção dos outliers
            outliers = (df_sem_outliers['Vegetation_Type'] == classe) & ((df_sem_outliers[coluna] < lower_bound) | (df_sem_outliers[coluna] > upper_bound))
            df_sem_outliers = df_sem_outliers[~outliers]
            
    # Return do dataset sem outliers
    return df_sem_outliers
```

```python
# Classes objetivo
atributo_objetivo = ['Type_2', 'Type_5', 'Type_6']

# Aplicação da função para remover os outliers das combinações da classe objetivo e atributos preditivos
df_all_sem_outliers = remover_outliers(df_all, atributo_objetivo, numericFeatures)
```

## Aplicação de Métodos
#### Logistic Regression (LR), Linear Discriminant Analysis (LDA) and Quadratic Discriminant Analysis (QDA)

```python
X = df_encoded.drop(columns=['Vegetation_Type', 'Vegetation_Type_Encoded'])
y = df_encoded['Vegetation_Type_Encoded']
```

#### Métodos: LR, LDA e QDA -> estes serão usados ao longo da implementação dos diferentes métodos de resampling

```python
#Métodos: LR, LDA and QDA:
from sklearn.linear_model import LogisticRegression as LR
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA

# Logistic Regression
logrLR = LR(multi_class='ovr') # ovr -> One Versus Rest
 
# Linear Discriminant Analysis
logrLDA= LDA()

# Quadratic Discriminant Analysis
logrQDA = QDA()
```

### Resampling Method: Holdout 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42, stratify=y)

#Verificar as condições no treino e teste 
print('Train size: ', X_train.shape)
print('Test size: ', X_test.shape)

# Converter os arrays NumPy para DataFrame para poder mostrar a distribuição da "Vegetation_Type" nos dados de treino e teste
df_train = pd.DataFrame({'condition': y_train})
df_test = pd.DataFrame({'condition': y_test})

# Exibir a distribuição das condições no treino e teste
print("\nTrain condition distribution:\n", df_train['condition'].value_counts(normalize=True))
print("\nTest condition distribution:\n", df_test['condition'].value_counts(normalize=True))
```

##### Neste caso, 70% dos dados são usados para treino (X_train e y_train), e 30% para teste (X_test e y_test). A estratificação é aplicada para garantir que a distribuição das classes seja mantida nos conjuntos de treino e teste, o que é particularmente útil, visto que, exite um desiquilibrio entre as classes.

#### Normalização dos dados

##### A normalização dos dados é um passo fundamental em muitos casos, especialmente quando utilizamos algoritmos que são sensíveis à escala de variáveis. <br> Todas as variáveis são reescaladas para ter média 0 e desvio padrão 1. <br> No nosso caso ao usarmos LR, LDA e QDA, a normalização é recomendada, estes algoritmos beneficiam deste processo e resultam em previsões mais precisas e estáveis.


```python
from sklearn.preprocessing import StandardScaler

# Inicializar o StandardScaler
scaler = StandardScaler()

#Normalizar o X de treino
Xstd_train = scaler.fit_transform(X_train)

#Normalizar o X de teste
Xstd_test = scaler.transform(X_test)
```

#### Avaliação do modelo

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

# Treinar os modelos
logrLR.fit(Xstd_train, y_train)
logrLDA.fit(Xstd_train, y_train)
logrQDA.fit(Xstd_train, y_train)

##########################LR##########################
# Previsões no conjunto de teste
y_pred_LR = logrLR.predict(Xstd_test)

# Avaliação do desempenho do modelo
classification_report_LR = classification_report(y_test, y_pred_LR)
confusion_matrix_LR = confusion_matrix(y_test, y_pred_LR, labels=[0, 1, 2])

# Converter a matriz para um DataFrame para mostrar classes
confusion_matrix_LR_df = pd.DataFrame(confusion_matrix_LR, 
                              index=['Type_2', 'Type_5', 'Type_6'], 
                              columns=['Pred_Type_2', 'Pred_Type_5', 'Pred_Type_6'])

print('\n############### Logistic Regression ###############')
print('Relatório de Classificação:\n', classification_report_LR)
print('Matriz de Confusão:\n', confusion_matrix_LR_df)

##########################LDA##########################
# Previsões no conjunto de teste
y_pred_LDA = logrLDA.predict(Xstd_test)

# Avaliação do desempenho do modelo
classification_report_LDA = classification_report(y_test, y_pred_LDA)
confusion_matrix_LDA = confusion_matrix(y_test, y_pred_LDA, labels=[0, 1, 2])

# Converter a matriz para um DataFrame para mostrar classes
confusion_matrix_LDA_df = pd.DataFrame(confusion_matrix_LDA, 
                              index=['Type_2', 'Type_5', 'Type_6'], 
                              columns=['Pred_Type_2', 'Pred_Type_5', 'Pred_Type_6'])

print('\n############### Linear Discriminant Analysis ###############')
print('Relatório de Classificação:\n', classification_report_LDA)
print('Matriz de Confusão:\n', confusion_matrix_LDA_df)

##########################QDA##########################
# Previsões no conjunto de teste
y_pred_QDA = logrQDA.predict(Xstd_test)

# Avaliação do desempenho do modelo
classification_report_QDA = classification_report(y_test, y_pred_QDA)
confusion_matrix_QDA = confusion_matrix(y_test, y_pred_QDA, labels=[0, 1, 2])

# Converter a matriz para um DataFrame para mostrar classes
confusion_matrix_QDA_df = pd.DataFrame(confusion_matrix_QDA, 
                              index=['Type_2', 'Type_5', 'Type_6'], 
                              columns=['Pred_Type_2', 'Pred_Type_5', 'Pred_Type_6'])

print('\n############### Quadratic Discriminant Analysis ###############')
print('Relatório de Classificação:\n', classification_report_QDA)
print('Matriz de Confusão:\n', confusion_matrix_QDA_df)
```

### Resampling Method: K-fold Cross Validation 

Como neste método fazemos várias divisões do datasample, é importante garantir que a distribuição para cada label é semelhante (uma vez que não o é no conjunto de dados originais), daí a utilização do StratifiedKFold e não do Kfold comum.

```python
# Normalizar os dados
scaler = StandardScaler() # Inicializar o StandardScaler
Xstd = scaler.fit_transform(X) # Normalizar toda a amostra X, posteriormente dividida em treino/teste com o método K-Fold Cross Validation
```

### Resampling Method: Bootstrap

```python
from sklearn.metrics import accuracy_score

n_samples = len(X)  # Total de amostra do dataset
bootstrap_range = [25,50,100] # Número de iterações
lr_accuracies_means = []
lr_accuracies_stds = []

# For cycle sobre o bootstrap_range
for n in bootstrap_range:
    print(f"Bootstrap iteration with {n} samples:")

    lr_accuracies_n = []

    # Realizar 'n' iterações de bootstrap
    for i in range(n):
        # Criar a amostra com bootstrapping (with replacement)
        train_indices = np.random.choice(n_samples, size=n_samples, replace=True)
        X_train_set = X.iloc[train_indices]
        Y_train_set = y.iloc[train_indices]
        
        # Criar a amostra de teste (os que não foram selecionados)
        test_indices = list(set(range(n_samples)) - set(train_indices))
        X_test_set = X.iloc[test_indices]
        y_test_set = y.iloc[test_indices]

        # Treinar o modelo de regressão logística
        logrLR.fit(X_train_set, Y_train_set)
        
        # Fazer previsões no conjunto de teste
        y_pred_lr = logrLR.predict(X_test_set)
        
        # Calcular a Accuracy
        lr_accuracy = accuracy_score(y_test_set, y_pred_lr)
        lr_accuracies_n.append(lr_accuracy)
        
    # Guardar a média e desvio padrão das Accuracies para cada n
    lr_accuracies_means.append(np.mean(lr_accuracies_n))
    lr_accuracies_stds.append(np.std(lr_accuracies_n))

plt.plot(bootstrap_range, lr_accuracies_means, label="Mean Accuracy")
plt.plot(bootstrap_range, lr_accuracies_stds, label="Std Deviation", linestyle="--")
plt.xlabel("Number of Bootstrap Samples")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
n_samples = len(Xstd)  # Total de amostra do dataset

# Definir o número de bootstrap samples
n_bootstrap = 25

# Criar listas para guardar os valores de accuracy e matrizes de confusão
lr_accuracies = []
lda_accuracies = []
qda_accuracies = []

lr_confusion_matrices = []
lda_confusion_matrices = []
qda_confusion_matrices = []

lr_reports = []
lda_reports = []
qda_reports = []

# Converter o conjunto de dados previamente normalizado (para o K-Fold Cross Validation) num dataframe
Xstd_df = pd.DataFrame(Xstd, index=X.index, columns=X.columns)

# Fazer loop a todas as bootstrap samples
for i in range(n_bootstrap):
    # Criar a amostra com bootstrapping (with replacement)
    train_indices = np.random.choice(n_samples, size=n_samples, replace=True)
    X_train_set = Xstd_df.iloc[train_indices]
    Y_train_set = y.iloc[train_indices]
    
    # Criar a amostra de teste (os que não foram selecionados)
    test_indices = list(set(range(n_samples)) - set(train_indices))
    X_test_set = Xstd_df.iloc[test_indices]
    y_test_set = y.iloc[test_indices]
    
    # Logistic Regression (LR)
    logrLR.fit(X_train_set, Y_train_set)
    y_pred_lr = logrLR.predict(X_test_set)
    lr_accuracy = accuracy_score(y_test_set, y_pred_lr)
    lr_accuracies.append(lr_accuracy)
    lr_report = classification_report(y_test_set, y_pred_lr, target_names=['Type_2', 'Type_5', 'Type_6'], output_dict=True)
    lr_reports.append(lr_report)
    lr_confusion_matrices.append(confusion_matrix(y_test_set, y_pred_lr, labels=[0, 1, 2]))

    # Linear Discriminant Analysis (LDA)
    logrLDA.fit(X_train_set, Y_train_set)
    y_pred_lda = logrLDA.predict(X_test_set)
    lda_accuracy = accuracy_score(y_test_set, y_pred_lda)
    lda_accuracies.append(lda_accuracy)
    lda_report = classification_report(y_test_set, y_pred_lda, target_names=['Type_2', 'Type_5', 'Type_6'], output_dict=True)
    lda_reports.append(lda_report)
    lda_confusion_matrices.append(confusion_matrix(y_test_set, y_pred_lda, labels=[0, 1, 2]))
    
    # Quadratic Discriminant Analysis (QDA)
    logrQDA.fit(X_train_set, Y_train_set)
    y_pred_qda = logrQDA.predict(X_test_set)
    qda_accuracy = accuracy_score(y_test_set, y_pred_qda)
    qda_accuracies.append(qda_accuracy)
    qda_report = classification_report(y_test_set, y_pred_qda, target_names=['Type_2', 'Type_5', 'Type_6'], output_dict=True)
    qda_reports.append(qda_report)
    qda_confusion_matrices.append(confusion_matrix(y_test_set, y_pred_qda, labels=[0, 1, 2]))

# Calcular a média e o desvio padrão de accuracy para cada modelo
avg_lr_accuracy = np.mean(lr_accuracies)
std_lr_accuracy = np.std(lr_accuracies)

avg_lda_accuracy = np.mean(lda_accuracies)
std_lda_accuracy = np.std(lda_accuracies)

avg_qda_accuracy = np.mean(qda_accuracies)
std_qda_accuracy = np.std(qda_accuracies)

# Função para calcular a média de um classification report
def average_classification_report(reports):
    # Inicializar o dicionário
    avg_report = {
        'Type_2': {'precision': 0, 'recall': 0, 'f1-score': 0, 'support': 0},
        'Type_5': {'precision': 0, 'recall': 0, 'f1-score': 0, 'support': 0},
        'Type_6': {'precision': 0, 'recall': 0, 'f1-score': 0, 'support': 0},
        'macro avg': {'precision': 0, 'recall': 0, 'f1-score': 0, 'support': 0},
        'weighted avg': {'precision': 0, 'recall': 0, 'f1-score': 0, 'support': 0}
    }
    
    # Somar os reports
    for report in reports:
        for key in avg_report.keys():
            for metric in ['precision', 'recall', 'f1-score', 'support']:
                avg_report[key][metric] += report[key][metric]
    
    # Calcular a média das métricas
    num_reports = len(reports)
    for key in avg_report.keys():
        for metric in ['precision', 'recall', 'f1-score', 'support']:
            avg_report[key][metric] /= num_reports

    return avg_report

# Calcular a média das matrizes de confusão e classification reports
avg_lr_confusion_matrix = np.mean(lr_confusion_matrices, axis=0)
avg_lda_confusion_matrix = np.mean(lda_confusion_matrices, axis=0)
avg_qda_confusion_matrix = np.mean(qda_confusion_matrices, axis=0)

avg_lr_report = average_classification_report(lr_reports)
avg_lda_report = average_classification_report(lda_reports)
avg_qda_report = average_classification_report(qda_reports)

print(f'Average Logistic Regression Accuracy: {avg_lr_accuracy:.4f}, Standard Deviation: {std_lr_accuracy:.4f}')
print(f'Average LDA Accuracy: {avg_lda_accuracy:.4f}, Standard Deviation: {std_lda_accuracy:.4f}')
print(f'Average QDA Accuracy: {avg_qda_accuracy:.4f}, Standard Deviation: {std_qda_accuracy:.4f}')

print(f'Average Logistic Regression Confusion Matrix:\n{avg_lr_confusion_matrix}')
print(f'Average LDA Confusion Matrix:\n{avg_lda_confusion_matrix}')
print(f'Average QDA Confusion Matrix:\n{avg_qda_confusion_matrix}')

print(f'Average Logistic Regression Classification Report:\n{avg_lr_report}')
print(f'Average LDA Classification Report:\n{avg_lda_report}')
print(f'Average QDA Classification Report:\n{avg_qda_report}')
```

### Ridge Regression

```python
from sklearn.model_selection import cross_val_predict, cross_val_score, KFold
from sklearn.metrics import classification_report, confusion_matrix

# Definir valores de alpha a testar
alpha_ridge = [1e-8, 1e-4, 1e-3, 1e-2, 1, 5, 10, 20, 25]

# Lista para armazenar os resultados da accuracy
accuracy_values = []

# Inicializar K-Fold Cross Validation com k=10
kf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

for alpha in alpha_ridge:
    # Inicializa o modelo RidgeClassifier com o valor de alpha atual
    ridge_model = LR(penalty='l2', C=1/alpha, solver='liblinear')
    
    # Aplicar cross-validation e calcular a média da accuracy
    accuracies = cross_val_score(ridge_model, Xstd, y, cv=kf, scoring='accuracy')
    accuracy_mean = np.mean(accuracies)
    
    # Armazenar a média de accuracy
    accuracy_values.append(accuracy_mean)

# Encontrar a maior accuracy e o índice correspondente
max_accuracy = max(accuracy_values)
max_accuracy_index = accuracy_values.index(max_accuracy)
best_alpha = alpha_ridge[max_accuracy_index]
print(f"Maior Accuracy: {max_accuracy:.3f} para alpha = {best_alpha}")

# Plotar o desempenho do modelo para diferentes valores de alpha
plt.figure(figsize=(10, 6))
sns.lineplot(x=np.log10(alpha_ridge), y=accuracy_values, marker='o', palette="dark")
plt.title('Desempenho do modelo Ridge para os diferentes valores de Alpha (Cross-Validation K=10)')
plt.xlabel('log10(alpha)')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

```python
# Inicializar o modelo RidgeClassifier com o melhor alpha
best_ridge_model = LR(penalty='l2', C=1/best_alpha, solver='liblinear')

# Usar Cross-Validation com K=10 para fazer previsões
kf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Fazer as previsões com K-Fold
y_pred = cross_val_predict(best_ridge_model, Xstd, y, cv=kf)

# Calcular a accuracy média utilizando cross_val_score
mean_accuracy = cross_val_score(best_ridge_model, Xstd, y, cv=kf, scoring='accuracy').mean()
print(f"Mean Accuracy: {mean_accuracy:.3f}")

# Calcular a matriz de confusão
conf_matrix = confusion_matrix(y, y_pred, labels=[0, 1, 2])

# Calcular o relatório de classificação (precisão, recall, f1-score)
class_report = classification_report(y, y_pred)

# Mostrar a matriz de confusão e o relatório de classificação
conf_matrix_df = pd.DataFrame(conf_matrix, 
                              index=['True_Type_2', 'True_Type_5', 'True_Type_6'], 
                              columns=['Pred_Type_2', 'Pred_Type_5', 'Pred_Type_6'])

print("\nRelatório de Classificação:\n", class_report)
print("Matriz de Confusão:\n", conf_matrix_df)

# Treinar o modelo no conjunto completo para aceder aos coeficientes
best_ridge_model.fit(Xstd, y)

# Aceder aos coeficientes do modelo treinado
coefficients = best_ridge_model.coef_[0]

# Criar um DataFrame para visualização dos coeficientes
coef_df = pd.DataFrame({'Atributo': X.columns, 'Coeficiente': coefficients})

# Ordenar os coeficientes em ordem crescente
sorted_coef_df = coef_df.sort_values(by='Coeficiente')

# Exibir os atributos com menor peso
print("Atributos com Menor Peso:")
print(sorted_coef_df.head(10))

# Plotar os coeficientes
plt.figure(figsize=(12, 8))
sns.barplot(x='Coeficiente', y='Atributo', data=sorted_coef_df, palette='viridis')
plt.title('Coeficientes do Modelo Ridge')
plt.xlabel('Coeficiente')
plt.ylabel('Atributo')
plt.show()
```

### Elastic Net Regression

```python
from sklearn.linear_model import ElasticNetCV, LogisticRegression

# Define os valores de alpha e l1_ratio a serem testados
alphas = [0.001, 0.01, 0.1, 1, 10, 50]
l1_ratios = [0.1, 0.5, 0.7, 0.9, 0.95]

# Inicializa o modelo ElasticNetCV para procurar os melhores valores de alpha e l1_ratio com K=10
elastic_cv = ElasticNetCV(alphas=alphas, l1_ratio=l1_ratios, cv=10, max_iter=1000, random_state=42)

# Fit ElasticNetCV
elastic_cv.fit(Xstd, y)

# Guarda os melhores valores encontrados
best_alpha = elastic_cv.alpha_
best_l1_ratio = elastic_cv.l1_ratio_
print(f"Melhor Alpha: {best_alpha}, Melhor L1_ratio: {best_l1_ratio}")

# Inicializar o modelo de Regressão Logística com elasticnet e os melhores parâmetros encontrados
best_elastic_model = LogisticRegression(penalty='elasticnet', solver='saga', C=1/best_alpha, l1_ratio=best_l1_ratio, max_iter=1000)

# Cross-Validation com K=10 para fazer previsões e calcular a média da accuracy
kf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

mean_accuracy = cross_val_score(best_elastic_model, Xstd, y, cv=kf, scoring='accuracy').mean()
print(f"Mean Accuracy: {mean_accuracy:.3f}")

# Previsões com Cross-Validation para calcular a matriz de confusão e o relatório de classificação
y_pred = cross_val_predict(best_elastic_model, Xstd, y, cv=kf)

# Matriz de Confusão
conf_matrix = confusion_matrix(y, y_pred, labels=np.unique(y))
conf_matrix_df = pd.DataFrame(conf_matrix, 
                              index=[f'True_{label}' for label in np.unique(y)], 
                              columns=[f'Pred_{label}' for label in np.unique(y)])

# Relatório de Classificação
class_report = classification_report(y, y_pred)
print("\nRelatório de Classificação:\n", class_report)
print("Matriz de Confusão:\n", conf_matrix_df)

# Treinar o modelo com todo o conjunto de dados para aceder aos coeficientes
best_elastic_model.fit(Xstd, y)
coefficients = best_elastic_model.coef_[0]

# Visualização dos Coeficientes
coef_df = pd.DataFrame({'Atributo': X.columns, 'Coeficiente': coefficients})
sorted_coef_df = coef_df.sort_values(by='Coeficiente')

# Exibir os atributos com menor peso
print("Atributos com Menor Peso:")
print(sorted_coef_df.head(10))

# Plotar os coeficientes
plt.figure(figsize=(12, 8))
sns.barplot(x='Coeficiente', y='Atributo', data=sorted_coef_df, palette='viridis')
plt.title('Coeficientes do Modelo ElasticNet')
plt.xlabel('Coeficiente')
plt.ylabel('Atributo')
plt.show()
```

**b) Tune the Random Forest hyperparameters, ensuring that your model is not overfitting
 the training data.**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import classification_report

# Grid de parametros
param_grid = {
    'n_estimators': [50, 100, 150, 200],
    'max_depth': [5, 10,  None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'],
    'bootstrap': [True, False]
}

random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_grid,
    n_iter=50,  # Numero de iterações
    scoring='f1_weighted',  # Metrica de otimização
    cv=cv,
    verbose=2,
    random_state=42,
    n_jobs=-1
)

# Perform the search
random_search.fit(Xstd_reduced, y)

# Best parameters and classification report
best_params = random_search.best_params_
best_model = random_search.best_estimator_

# Evaluate the best model with cross-validation
y_pred = cross_val_predict(best_model, Xstd_reduced, y, cv=cv)
class_report = classification_report(y, y_pred)

print("Best Parameters:")
print(best_params)
print("\nClassification Report:")
print(class_report)
```
