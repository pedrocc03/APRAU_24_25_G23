### Features:

* **Altitude**: A altitude do terreno em metros.
* **Slope_Orientation (Orientação da Inclinação)**: A direção da bússola para a qual o terreno está voltado (em graus).
* **Slope (Inclinação)**: A inclinação em graus.
* **Horizontal_Distance_To_Water (Distância Horizontal até a Água)**: Distância horizontal até a superfície de água mais próxima (em metros).
* **Vertical_Distance_To_Water (Distância Vertical até a Água)**: Distância vertical até a superfície de água mais próxima (em metros).
* **Horizontal_Distance_To_Roadways (Distância Horizontal até Estradas)**: Distância horizontal até a estrada mais próxima, em metros (pode indicar influência de atividades humanas no ambiente).
* **Shadow_Index (Índice de Sombra)**: Índice de sombra medido em três horários do dia (9h, 12h e 15h).
* **Horizontal_Distance_To_Fire_Points (Distância Horizontal até Pontos de Fogo)**: Distância horizontal até os pontos de ignição de incêndios mais próximos (em metros).
* **Canopy_Density (Densidade do Dossel)**: A percentagem da área do terreno coberta por dossel de árvores (medido em percentagem, de 0 a 100).
* **Rainfall_Summer (Precipitação no Verão)**: Precipitação média durante o verão (medida em mm).
* **Rainfall_Winter (Precipitação no Inverno)**: Precipitação média durante o inverno (medida em mm).
* **Wind_Exposure_Level (Nível de Exposição ao Vento)**: Velocidade média do vento durante o inverno (medida em km/h).
* **Soil_Type (Tipo de Solo)**: Designação do tipo de solo. No conjunto de dados existem 40 tipos diferentes de solo.
* **Wilderness_Area (Área Selvagem)**: No conjunto de dados existem 4 tipos diferentes de áreas selvagens.
* **Vegetation_Type (Tipo de Vegetação)**: A variável alvo. Existem 7 classes possíveis, cada uma representando um tipo diferente de vegetação.

```python
# Função para verificar NaNs e Outliers
def check_nans_outliers(df, class_name):
    print(f"Análise de NaNs e Outliers para {class_name}")
    
    # Filtrar para a classe objetivo
    df_class = df[df['Vegetation_Type'] == class_name] 
    
    # Verificar NaNs
    print(f"\n{class_name} - Verificação de NaNs:")
    print(df_class.isnull().sum())
    
    # Selecionar apenas colunas numéricas
    df_numeric = df_class.select_dtypes(include=[np.number])
    
    # Análise de Outliers
    print(f"\n{class_name} - Outliers (nas colunas numéricas):")
    Q1 = df_numeric.quantile(0.25)
    Q3 = df_numeric.quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df_numeric < (Q1 - 1.5 * IQR)) | (df_numeric > (Q3 + 1.5 * IQR)))
    
    # Mostrar o número de outliers em cada coluna
    print(outliers.sum())

# Verificação de NaNs e Outliers para as 3 classes
for class_name in df_all['Vegetation_Type'].unique():
    check_nans_outliers(df_all, class_name)
```
